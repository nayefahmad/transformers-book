```python
import llm
from kaggle_secrets import UserSecretsClient

user_secrets = UserSecretsClient()
secret_value_0 = user_secrets.get_secret("test")
```

```python
model = llm.get_model("gpt-3.5-turbo")
model
```

```python
model.key = secret_value_0
```

```python
def create_question(text):
    prompt = f"""
    Read the following sentence and answer the question. 

    My cats' names are Hobbes and Roosey. 

    Question: {text} 
    """
    return prompt


q_01 = create_question("How many letters are in the names of each of my cats?")
q_02 = create_question("Can you come up with similar cat names, that start with the same letters?")
```

```python
model.prompt(q_01).text()
```

```python
model.prompt(q_02).text()
```
